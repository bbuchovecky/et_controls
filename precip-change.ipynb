```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

import xclimate as xclim
```

```python
nonglc_pct_threshold = 80
grid = xclim.load_fhist_ppe_grid()
nonglc_mask = grid.PCT_GLC < nonglc_pct_threshold
```

```python
variables = ["PRECC", "PRECL", "uIVT", "vIVT", "TMQ", "U", "V"]
time_slice = slice("1995-01", "2014-12")

fhist = {}
for v in variables:
    print(v)
    key = f"{v}_month_1"
    fhist[key] = xclim.load_fhist(key, keep_var_only=True)[v].sel(time=time_slice)
    if "lev" in fhist[key].dims:
        fhist[key] = fhist[key].isel(lev=-1)
```
